```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
import glob
import os
```

```python
files = glob.glob("./test/predictions/*")
```

```python
file = files[3]
```

```python
img = cv2.imread(file, 0)
```

```python
img = np.where(img>128, 255, 0 ).astype(np.uint8)

plt.imshow(img, cmap="gray")
```

```python
img_erode = np.zeros_like(img)
img_dil = np.zeros_like(img)
img_new = np.zeros_like(img)
kernel = np.ones((5,5))
```

```python
cv2.erode(img, np.ones((3,3)), img_erode, iterations=2)
cv2.dilate(img_erode, np.ones((7,7)), img_dil, iterations=4)
cv2.erode(img_dil, np.ones((3,3)), img_new, iterations=6)
```

```python
plt.imshow(img_new, cmap="gray")
```
